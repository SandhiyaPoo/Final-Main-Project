## Stage 2 – Data Cleaning and Pre-processing

**Handle Missing Values(Impute or Drop)**

```python
# For Numerical Columns
# Fill Numerical Missing Values With Mean

num_cols = df.select_dtypes(include=['int64','float64'] )
df[num_cols.columns] = num_cols.fillna(num_cols.mean())
```

```python
# For Categorical Columns
# Fill Categorical Missing Values With Mode

cat_cols = df.select_dtypes(include='object')

for col in cat_cols:
     df[col].fillna(df[col].mode()[0],
inplace=True)



```

```python
# Drop Rows With Missing Values

df.dropna(inplace=True)
```

**Handle Duplicates**

```python
# Remove Duplicates

df.drop_duplicates(inplace=True)
```

**Rename Columns or Drop Unwanted Columns**

```python
# Rename Columns

df.rename(columns={
    'sales_amount':'Sales',
    'customer_name':'Customer_Name'
}, inplace=True)
```

```python
# Drop Unwanted Columns

df.drop(columns=['Unnamed: 0'],
inplace=True, errors='ignore')
```

**Treat Outliers if Required**

```python
# Visualize Outliers

import matplotlib.pyplot as plt

plt.boxplot(df['Sales_Amount'])
plt.title("Boxplot of Sales")
plt.show()
```

```python
# Remove Outliers Using IQR Method

Q1 = df['Sales_Amount'].quantile(0.25)
Q3 = df['Sales_Amount'].quantile(0.75)

IQR = Q3-Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df = df[(df['Sales_Amount'] >= lower_limit)& (df['Sales_Amount'] <= upper_limit)]
```

**Check Skewness and Apply Transformations**

```python
# Check Skewness
# Check skewness of numerical columns

df.skew(numeric_only=True)
```

```python
# Apply Log Transformation

df['Sales_Amount'] = np.log1p(df['Sales_Amount'])
```

**Convert Data Types if Needed**

```python
# Check Data Types

df.dtypes
```

```python
# Convert Date Column
# Convert Order Date Into Datetime Format

df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

**Feature Transformations**

```python
# Extract Month and Year

df['Month'] = df['Order_Date'].dt.month
df['Year'] = df['Order_Date'].dt.year
```

```python
# Create Derived Field
# Create Total Amount column

df['Total_Sales'] = df['Unit_Price'] * df['Units_Sold']
```

```python
# Convert city names into lower

df['City'] = df['City'].str.lower()
```

```python
# Replace with proper city name

df['City'] = df['City'].replace({
    'hyd':'hyderabad',
    'bangalore':'bengaluru'
})

# Convert first letter of each word into capital
df['City'] = df['City'].str.title()
```

```python
# Calculate Sales Amount

df['Sales_Amount'] = (
    df['Units_Sold'] * df['Unit_Price']
) * (1 - df['Discount_%'])
```

```python
# Convert decimal discount into percentage

df['Discount_%'] = df['Discount_%'] * 100
```

```python
# Convert negative values into positive

df['Sales_Amount'] = df['Sales_Amount'].abs()

df['Profit'] = df['Profit'].abs()
```

```python
# Keep discount values less than or equal to 100%

df = df[df['Discount_%'] <= 100]
```

```python
# Replace zero values with Mean

cols = ['Units_Sold', 'Sales_Amount', 'Total_Sales']
for col in cols:
   df[col] = df[col].replace(0,
df[col].mean())
```

```python
# Round Numerical columns

df = df.round(2)
```

### Analys abaout the sales by region

```python
df['Region'] = df['City'].replace({
    'Delhi':'North',
    'Mumbai':'West',
    'Ahmedabad':'West',
    'Pune':'West',
    'Bengaluru':'South',
    'Hyderabad':'South'
})
```

```python
sales_region = df.groupby('Region')['Sales_Amount'].sum().reset_index()

print(sales_region)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.barplot(
    x='Region',
    y='Sales_Amount',
    data=sales_region
    )

plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales Amount')

plt.show()
```

### **Insights**

* The chart campares total sales across North,South and West region.

* It help identify which region generates the highest revenue.

* It shows regional sales performance and business contribution.

* Regions with lower sales indicate opportunities for improvement.
